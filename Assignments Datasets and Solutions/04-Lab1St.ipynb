###Step 14: 
Build your K-neighbors classifier for nearest neighbor of 1 using from sklearn.neighbors import KNeighborsClassifier, fit the model to your train dataset and make a prediction for teh data point of [5, 1.9, 1, 0.2]. Print your prediction class value as an integer and also the corresponding string label.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
import numpy as np

# Step 1: Load iris dataset from ScikitLearn
iris_dataset = load_iris()
X = iris_dataset.data
y = iris_dataset.target

# Step 12: Split dataset into train and test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# Step 14: Build K-neighbors classifier for nearest neighbor of 1
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

# Make a prediction for the data point [5, 1.9, 1, 0.2]
X_new = np.array([[5, 1.9, 1, 0.2]])
prediction = knn.predict(X_new)
print("Prediction class value: ", prediction[0])
print("Corresponding string label: ", iris_dataset.target_names[prediction[0]])
```

we use the KNeighborsClassifier class from scikit-learn to build a k-nearest neighbors model with a value of n_neighbors set to 1. This means that the model will predict the class of a new data point based on the class of its nearest neighbor in the training set.

We then fit the model to our training data using the fit method of the KNeighborsClassifier class.

Next, we create a new data point X_new containing the values [5, 1.9, 1, 0.2], which represents a new iris flower that we want to predict the species of.

We then use the predict method of the KNeighborsClassifier class to make a prediction for this new data point based on the trained model. The predicted class value is stored in the prediction variable.

Finally, we print the predicted class value and the corresponding string label for the predicted iris species using the target_names attribute of the iris_dataset object.

In this case, the predicted class value is 0, which corresponds to the 'setosa' species of iris, and this is confirmed by printing the corresponding string label as 'setosa'.

###Step 10
Visualize the relationships between these data points.

Create a function to scale your dataset by using the formula 𝑏=𝑥−𝑚𝑖𝑛𝑚𝑎𝑥−𝑚𝑖𝑛

Using this function, scale displacement, horsepower, acceleration, weight, andmpg`.
Create a boxplot of mpg for different origin values before and after scaling.

```python
def scale_data(df):
    """
    This function scales the continuous columns of the DataFrame using the formula:
    b = (x - xmin) / (xmax - xmin)
    """
    continuous_cols = ['displacement', 'horsepower', 'acceleration', 'weight', 'mpg']
    
    for col in continuous_cols:
        xmin = df[col].min()
        xmax = df[col].max()
        df[col] = (df[col] - xmin) / (xmax - xmin)
    
    return df
```

To visualize the relationships between the features, we use a scatter plot matrix. This will allow us to see how the features are correlated with each other.

To scale the dataset, we define a function that takes a pandas DataFrame as input and returns a scaled DataFrame.

```python
import seaborn as sns

# Boxplot of mpg for different origin values before scaling
sns.boxplot(x='origin', y='mpg', data=auto_data)
plt.title('Boxplot of mpg for different origin values before scaling')

# Scale the data
auto_data_scaled = scale_data(auto_data)

# Boxplot of mpg for different origin values after scaling
sns.boxplot(x='origin', y='mpg', data=auto_data_scaled)
plt.title('Boxplot of mpg for different origin values after scaling')
```

we create a boxplot of mpg for different origin values before and after scaling using seaborn.

The first boxplot shows the distribution of mpg for different origin values before scaling. The second boxplot shows the distribution of mpg for different origin values after scaling.

The scaled boxplot is more informative because it shows the range of values for mpg for each origin group. It also shows that there is less variability in mpg after scaling.
